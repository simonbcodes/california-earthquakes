```python
from wikitables import import_tables
import pandas as pd
from bokeh.plotting import figure, output_notebook, show, ColumnDataSource

output_notebook()
```

```python
wiki_tables = import_tables('List of earthquakes in California') # load data
```

```python
dataset = pd.read_json(wiki_tables[0].json())[:-1] # read into df and drop bottom row

dataset.at[75, 'Date'] = '1838-06-01' # set specfic value

dataset['Date'] = dataset.apply(lambda row: pd.to_datetime(row['Date']), axis=1) # clean data
dataset['MMI'] = dataset.apply(lambda row: row['MMI'].split(' ')[0], axis=1)
dataset['Deaths'] = dataset['Deaths'].replace('', 0)
dataset['Injuries'] = dataset['Injuries'].replace('', 0)
dataset = dataset.drop('Total damage / notes', axis=1)

dataset
```

```python
plot = figure(title="Earthquake Intensity over Time (Mercalli Intensity)", 
              x_axis_label='Time',
              y_axis_label='Intensity',
              x_axis_type='datetime',
              y_range=(4, 13))

plot.scatter(dataset['Date'], dataset['MMI'])

show(plot)
```
